```python
from math import *
from numpy import *
import matplotlib.pyplot as plt
```

```python
def extraer_individual_curves_log(filename):
    datos = loadtxt(filename, float)
    Time=[]
    Flux=[]
    Time_low=[]
    Time_high=[]
    Flux_low=[]
    Flux_high=[]
    for i in range(0,len(datos)):
        Time.append(datos[i][0])
        Flux.append(datos[i][1])
        Time_low.append(datos[i][2])
        Time_high.append(datos[i][3])
        Flux_low.append(datos[i][4])
        Flux_high.append(datos[i][5])
#    name.append(str(datos[i]))
    return Time,Flux,Time_low,Time_high,Flux_low,Flux_high

def broken_power_law(x,t_break,alpha_1,alpha_2,amplitude):
    alpha = where(x< t_break, alpha_1, alpha_2)
    tt = x / t_break
    return amplitude * tt ** (-alpha)

def power_law(x,alpha_1,amplitude):
    return amplitude * x ** (-alpha_1)
```
