```python
def extraer_individual_curves_log(filename):
    datos = loadtxt(filename, float)
    Time=[]
    Flux=[]
    Time_low=[]
    Time_high=[]
    Flux_low=[]
    Flux_high=[]
    for i in range(0,len(datos)):
        Time.append(datos[i][0])
        Flux.append(datos[i][1])
        Time_low.append(datos[i][2])
        Time_high.append(datos[i][3])
        Flux_low.append(datos[i][4])
        Flux_high.append(datos[i][5])
#    name.append(str(datos[i]))
    return Time,Flux,Time_low,Time_high,Flux_low,Flux_high

def broken_power_law(x,t_break,alpha_1,alpha_2,amplitude):
    alpha = where(x< t_break, alpha_1, alpha_2)
    tt = x / t_break
    return amplitude * tt ** (-alpha)

def power_law(time,alpha_1,amplitude):
    tt = time
    return amplitude * tt ** (-alpha_1)
```

```python
from lmfit import Model
from scipy.optimize import curve_fit
import corner
import emcee


def broken_power_law(x,t_break,alpha_1,alpha_2,amplitude):
    alpha = where(x< t_break, alpha_1, alpha_2)
    tt = x / t_break
    return amplitude * tt ** (-alpha)

def power_law(x,alpha_1,amplitude):
    tt = x
    return amplitude * tt ** (-alpha_1)


def lnlikehood(x, y, yerr, theta,BPL):
    if(BPL==True):
        yerr=array(yerr)
        t_break,alpha_1,alpha_2,amplitude = theta
        model = broken_power_law(x,t_break,alpha_1,alpha_2,amplitude)
        return -0.5*(sum((y-model)**2/(yerr**2) + log(sqrt(yerr**2)*2*pi)))
    else:
        yerr=array(yerr)
        alpha_1,amplitude = theta
        model = power_law(x,alpha_1,amplitude)
        return -0.5*(sum((y-model)**2/(yerr**2) + log(np.sqrt(yerr**2)*2*pi)))

def information_criteria(ln_like,y,theta):
    AIC=-2*ln_like+(2*len(theta))+(2*len(theta)*len(theta))
    BIC=-2*ln_like+(len(theta)*log(len(y)))
    dof=len(y)-len(theta)
    return AIC,BIC,dof
```

```python
def lnlike(theta, x, y, yerr):
    t_break,alpha_1,alpha_2,amplitude = theta
    model = broken_power_law(x,t_break,alpha_1,alpha_2,amplitude)
    inv_sigma2 = 1.0/(yerr**2)
    return -0.5*(sum((y-model)**2*inv_sigma2 - log(inv_sigma2)))

def lnprior(theta):
    t_break,alpha_1,alpha_2,amplitude = theta
    if 1e1<t_break<1e4 and -1.0<alpha_1<0.2 and 0.0<alpha_2<4.0 and 1e-15<amplitude<1e-11:
        return 0.0
    return -inf

def lnprob(theta, x, y, yerr):
    lp = lnprior(theta)
    if not isfinite(lp):
        return -inf
    return lp + lnlike(theta, x, y, yerr)

def MCMC_fitting(time,lumin,err_tot,name):
    t_break_true,alpha_1_true,alpha_2_true,amplitude_true=mean(time),-1.0,1.0,max(lumin)
#    L_0_true,T_sp_true=max(np.log10(lumin)), np.log10(np.mean(time))
    import scipy.optimize as op
    nll = lambda *args: -lnlike(*args)

    time=array(time)
    lumin=array(lumin)
    err_tot=array(err_tot)

    result = op.minimize(nll, [t_break_true,alpha_1_true,alpha_2_true,amplitude_true], args=(time,lumin,err_tot))
#    L_0_ml,T_sp_ml = result["x"]
    t_break_ml,alpha_1_ml,alpha_2_ml,amplitude_ml = result["x"]

    ndim, nwalkers = 4, 20
    pos = [result["x"] + [mean(time),-1.0,1.0,max(lumin)]*random.randn(ndim) for i in range(nwalkers)]
    sampler = emcee.EnsembleSampler(nwalkers, ndim, lnprob, args=(time,lumin,err_tot))
    sampler.run_mcmc(pos, 5000)
    samples_1 = sampler.chain[:, 500:, :].reshape((-1, ndim))
    
    samples_2 = sampler.flatchain
    theta_max  = samples_2[argmax(sampler.flatlnprobability)]
    
    samples=[]
    for i in range(0,len(samples_1)):
        if(1e1<samples_1[i][0]<1e4 and -1.0<samples_1[i][1]<0.2 and 0.0<samples_1[i][2]<4.0 and 1e-15<samples_1[i][3]<1e-11):
            samples.append([samples_1[i][0],samples_1[i][1],samples_1[i][2],samples_1[i][3]])
    samples=array(samples)
    
    fig, axes = plt.subplots(4, figsize=(10, 7), sharex=True)
    #samples = sampler.get_chain()
    labels = ["$t_{break}$", "$\\tau_1$", "$\\tau_2$", "$A$"]
    for i in range(ndim):
        ax = axes[i]
        ax.plot(samples[:, i], "k", alpha=0.9)
        ax.set_xlim(0, len(samples))
        ax.set_ylabel(labels[i])
        ax.yaxis.set_label_coords(-0.1, 0.5)

    axes[-1].set_xlabel("step number");
    plt.savefig(name+'_running.pdf',format='pdf')
    plt.close()
    
    samples[:, 1] = (samples[:, 1])
    t_break_mcmc,alpha_1_mcmc,alpha_2_mcmc,amplitude_mcmc = map(lambda q: (q[1], q[0], q[2]),zip(*percentile(samples, [16, 50, 84],axis=0)))

    fig = corner.corner(samples, labels=["$t_{break}$", "$\\tau_1$", "$\\tau_2$", "$A$"], quantiles=[0.16, 0.5, 0.84],color='indigo',
                        linewidth=4.0,show_titles=True,plot_datapoints=False,fill_contours=True,bins=80, smooth=1.0)
    plt.savefig(name+'_corner.pdf',format='pdf')
    plt.close()
    
    t_break_MCMC=[]
    alpha_1_MCMC=[]
    alpha_2_MCMC=[]
    amplitude_MCMC=[]
    for t_break_0,alpha_1_0,alpha_2_0,amplitude_0 in samples[random.randint(len(samples), size=200)]:
        t_break_MCMC.append(t_break_0)
        alpha_1_MCMC.append(alpha_1_0)
        alpha_2_MCMC.append(alpha_2_0)
        amplitude_MCMC.append(amplitude_0)
        
    return t_break_mcmc,alpha_1_mcmc,alpha_2_mcmc,amplitude_mcmc,t_break_MCMC,alpha_1_MCMC,alpha_2_MCMC,amplitude_MCMC
```
